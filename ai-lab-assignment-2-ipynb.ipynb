```python
from sklearn.datasets import load_digits
```

```python
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
```

```python
print("Dataset shape: ", digits.data.shape)
print("Image shape: ", digits.images.shape)
```

```python
print(digits.keys())
```

```python
print("Target labels",digits.target_names)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(digits.images[5], cmap='gray')
plt.title(f"Label: {digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
flg, axes = plt.subplots(2, 5, figsize=(10,5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap ='gray')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for ax, image, target in zip(axes.flat, digits.images, digits.target):
    ax.imshow(image, cmap='magma')
    ax.set_title(f"Digit: {target}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
